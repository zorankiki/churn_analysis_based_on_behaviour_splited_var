```python
import significant_variables
import variables_with_p_below_0_2
import important_variables
import yaml
import pandas as pd
import read_a_combination_of_variables
import os
import sys
sys.path.insert(0, os.path.abspath('../'))
from helpers.s3_bucket_utils import S3BucketUtils
from helpers import settings

bucket = S3BucketUtils()

with open(r'./parameters/started_doing_something_report_parameters.yaml') as file:
    model_params = yaml.load(file, Loader=yaml.FullLoader)

date_of_analysis = model_params['date_of_analysis']
date_dir = date_of_analysis.replace('-', '_')

# churn_based_on_behaviour_dir = input('Name of the data directory: ')
churn_based_on_behaviour_dir = 'churn_analysis_based_on_behaviour/'

# p_limit = input('Used p limit: ').replace('.', '_')
p_limit = '0_2'

without_properly_used_vars = 'No'
if without_properly_used_vars == 'Yes':
    without_properly_used_vars = True
elif without_properly_used_vars == 'No':
    without_properly_used_vars = False

if without_properly_used_vars == True:
    model_numbers = \
    read_a_combination_of_variables.\
    get_a_list_of_model_numbers(dir_name='combinations_of_variables_that_are_not_dependent/combinations_and_names_without_properly_used_inquiries_vars/')
else:
    model_numbers = \
    read_a_combination_of_variables.\
    get_a_list_of_model_numbers(dir_name='combinations_of_variables_that_are_not_dependent/')


# exports_dir_path_from = 'exports/properly_used_inquiries/coefficients_and_pvalues'
# exports_dir_path_from = input('Read the exports from: ')
exports_dir_path_from = 'exports/coefficients_and_pvalues'
# exports_dir_path_to = 'exports/properly_used_inquiries'
# exports_dir_path_to = input('Write exports to: ')
exports_dir_path_to = 'exports'

for model_type in ['started_doing_something', 'stopped_doing_something']:
    for model_number in model_numbers[1:]:
        if 'model_'+str(model_number) in os.listdir('data/'+date_dir+'/'+exports_dir_path_from+'/'+model_type+'/'):
            sign_vars_export_sorted_by_pvalue = significant_variables.get_significant_variables(data_dir=churn_based_on_behaviour_dir, \
                                               date_dir=date_dir, exports_dir_path_from=exports_dir_path_from, exports_dir_path_to=exports_dir_path_to,\
                                                                                                model_type=model_type, model_number=model_number,\
                                                                                      p_limit=p_limit,sort_by = 'p_value')

            sign_vars_export_sorted_by_var_name = significant_variables.get_significant_variables(data_dir=churn_based_on_behaviour_dir, \
                                               date_dir=date_dir, exports_dir_path_from=exports_dir_path_from, exports_dir_path_to=exports_dir_path_to,\
                                                                                                  model_type=model_type, model_number=model_number,\
                                                                                        p_limit=p_limit,sort_by = 'variable_name')

            vars_with_p_below_0_2_export_sorted_by_pvalue = variables_with_p_below_0_2.get_significant_variables(data_dir=churn_based_on_behaviour_dir, \
                                               date_dir=date_dir, exports_dir_path_from=exports_dir_path_from, exports_dir_path_to=exports_dir_path_to, model_type=model_type, model_number=model_number, p_limit=p_limit,\
                                              sort_by = 'p_value')

            vars_with_p_below_0_2_export_sorted_by_var_name = variables_with_p_below_0_2.get_significant_variables(data_dir=churn_based_on_behaviour_dir, \
                                               date_dir=date_dir, exports_dir_path_from=exports_dir_path_from, exports_dir_path_to=exports_dir_path_to, model_type=model_type, model_number=model_number, p_limit=p_limit,\
                                              sort_by = 'variable_name')
important_variables.main(without_properly_used_vars=without_properly_used_vars, churn_based_on_behaviour_dir=churn_based_on_behaviour_dir, date_dir=date_dir,\
    exports_dir_path=exports_dir_path_to)        
```
